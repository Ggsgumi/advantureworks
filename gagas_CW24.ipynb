# Deskripsi Data untuk Tabel Database Penjualan

---

## Tabel `product`
Menyimpan informasi terkait produk yang dijual oleh perusahaan.

- **product_key** *(int)*: Identifier untuk setiap produk.
- **name** *(varchar)*: Nama produk.
- **standard_cost** *(varchar)*: Biaya standar produksi per produk (dalam dolar)
- **color** *(varchar)*: Warna produk.
- **subcategory** *(varchar)*: Subkategori dari produk. (contoh: Bottles and Cages, Touring Frames, Handlebars, Cleaners)
- **category** *(varchar)*: Kategori produk. (Bikes, Accessories, Clothing, Components)
- **background_color_format** *(varchar)*: Warna latar belakang produk.
- **font_color_format** *(varchar)*: Warna font prodok.

---

## Tabel `region`
Menyimpan informasi terkait wilayah penjualan produk

- **sales_teritory_key** *(int)*: Indentifier regionn.
- **name** *(varchar)*: Nama wilayah.
- **country** *(varchar)*: Negara tempat wilayah berada.
- **group** *(varchar)*: Grup wilayah, dapat berupa negara atau negara bagian.

---

## Tabel `reseller`
Menyimpan data tentang reseller yang menjual produk dari perusahaan.

- **reseller_key** *(int)*: Identifier untuk setiap reseller.
- **name** *(varchar)*: Nama reseller.
- **bussiness_type** *(varchar)*: Jenis bisnis yang dimiliki oleh reseller, seperti Warehouse, Specialty Bike Shop atau Value Added Reseller.
- **city** *(varchar)*: Kota tempat reseller berada.
- **state_province** *(varchar)*: Provinsi atau negara bagian dari reseller.
- **country** *(varchar)*: Negara tempat reseller.

---

## Tabel `sales_person`
Menyimpan data tentang karyawan yang bertanggung jawab atas departemen sales.

- **employee_key** *(int)*: Indentifier karyawan .
- **employee_id** *(int)*: ID unik dari karyawan di dalam perusahaan.
- **name** *(varchar)*: Nama karyawan.
- **title** *(varchar)*: Jabatan atau posisi pekerjaan.
- **upn** *(varchar)*: User Principal Name atau username yang digunakan sebagai alamat email karyawan (username@adventureworks.com)

---

## Tabel `sales`
Menyimpan catatan transaksi penjualan produk.

- **sales_key** *(int)*: Identifier unik untuk setiap transaksi penjualan.
- **sales_order_number** *(varchar)*: Nomor untuk setiap pesanan penjualan.
- **order_date** *(date)*: Tanggal pesanan dibuat.
- **product_key** *(int)*: Produk yang dijual, mengacu ke `product` kolom product_key.
- **reseller_key** *(int)*: Reseller yang membeli produk, mengacu ke `reseller` kolom reseller_key.
- **employee_key** *(int)*: Karyawan yang menangani penjualan, mengacu ke `sales_person` kolom employee_key.
- **sales_teritory_key** *(int)*: Wilayah transaksi penjualan, mengacu ke `region` kolom sales_teritory_key.
- **quantity** *(varchar)*: Jumlah produk yang dijual.
- **unit_price** *(varchar)*: Harga per unit produk. (dalam dolar)
- **sales** *(varchar)*: Total nilai penjualan untuk transaksi. (dalam dolar)
- **cost** *(varchar)*: Total biaya dari transaksi tersebut. (dalam dolar)

---

## Tabel `sales_person_region`
Menyimpan hubungan antara karyawan penjualan dan wilayah penjualan yang dikelola.

- **employee_key** *(int)*: FK karyawan, mengacu ke `sales_person`.
- **SalesTerritoryKey** *(int)*: FK wilayah penjualan, mengacu ke `region`.

---

## Tabel `target`
Menyimpan informasi tentang target penjualan bulanan yang diberikan kepada setiap karyawan.

- **employee_key** *(int)*: karyawan yang memiliki target, mengacu ke `sales_person`.
- **target_month** *(date)*: Bulan target penjualan.
- **target_value** *(varchar)*: Nilai target penjualan dalam bentuk angka. (dalam dolar)

---

```python
# Please load this library
import numpy as np
import pandas as pd
from datetime import date,datetime

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_product = pd.read_csv('/content/product.csv')
df_region = pd.read_csv('/content/region.csv')
df_reseller = pd.read_csv('/content/reseller.csv')
df_sales = pd.read_csv('/content/sales.csv')
df_sales_person = pd.read_csv('/content/sales_person.csv')
df_sales_person_region = pd.read_csv('/content/sales_person_region.csv')
df_target = pd.read_csv('/content/target.csv')
```

```python
# rename some column that have similar name
df_product = df_product.rename(columns={'name':'product_name'})
df_region = df_region.rename(columns={'name':'area','country':'country_area'})
df_reseller = df_reseller.rename(columns={'name':'reseller_name','country':'reseller_country'})
df_sales_person = df_sales_person.rename(columns={'name':'emprolyee_name','title':'employee_title','upn':'username'})
df_sales_person_region = df_sales_person_region.rename(columns={'SalesTerritoryKey':'sales_teritory_key'})
```

```python
# Join df_product and df_sales on product_key
df = pd.merge(df_product, df_sales, on='product_key', how='inner')
```

```python
# Join with df_reseller on reseller_key
df = pd.merge(df, df_reseller, on='reseller_key', how='inner')
```

```python
# Create second dataframe and Join beetween df_sales_person and df_target on employee_key
df_2 = pd.merge( df_sales_person,df_target, on='employee_key', how='inner')
```

```python
# Join second dataframe with df_sales_person_region on employee_key
df_2 = pd.merge(df_2, df_sales_person_region, on='employee_key', how='inner')
```

```python
# Join second dataframe with df_region on sales_teritory_key
df_2 = pd.merge(df_2, df_region, on='sales_teritory_key', how='inner')
```

```python
# Join between first and second dataframe on employee_key and sales_teritory_key
df_merge = pd.merge(df, df_2, on=['employee_key','sales_teritory_key'], how='inner')
```

```python
# show all columns in dataframe
pd.set_option("display.max_columns",None)
df_merge.head()
```

```python
df_merge[['standard_cost','unit_price','sales','cost','target_value']] = df_merge[['standard_cost','unit_price','sales','cost','target_value']].replace({'\$':'',',':''}, regex=True)
```

```python
df_merge[['standard_cost','unit_price','sales','cost','target_value']] = df_merge[['standard_cost','unit_price','sales','cost','target_value']].astype(float)
```

```python
df_merge['order_date'] = pd.to_datetime(df_merge['order_date'])
df_merge['target_month'] = pd.to_datetime(df_merge['target_month'])
```

```python
df_merge['order_date_year'] = df_merge['order_date'].dt.year
df_merge['order_date_month'] = df_merge['order_date'].dt.month

df_merge['target_month_by_years'] = df_merge['target_month'].dt.year
df_merge['target_month_by_month'] = df_merge['target_month'].dt.month
```

```python
count_category_order = df_merge.groupby('category').agg({'quantity':'sum'}).reset_index()
count_category_order.sort_values(by ='quantity', ascending=False)
```

```python
# barplot
plt.figure(figsize=(10, 3))
sns.barplot(x = 'category',
            y = 'quantity',
            hue = 'category',
            data = count_category_order.sort_values(by ='quantity', ascending=False))
plt.xlabel("Product Category")
plt.ylabel("Total Order")
plt.title("Total Number of Order by Product Category")
plt.show()
```

```python
count_subcategory_order = df_merge.groupby(['subcategory','category']).agg({'quantity':'sum'}).reset_index()
count_subcategory_order.sort_values(by ='category', ascending=False)
```

```python
# barplot
plt.figure(figsize=(10, 3))
sns.barplot(x = 'subcategory',
            y = 'quantity',
            hue = 'category',
            data = count_subcategory_order.sort_values(by ='category', ascending=False))
plt.xlabel("Subcategory")
plt.ylabel("Total Order")
plt.title("Total Number of Order by Subategory")
plt.xticks(rotation=90)
plt.xticks(fontsize=6)
plt.show()
```

```python
new_df = df_merge.copy()
```

```python
# create new column for profit
new_df['profit'] =  new_df['unit_price'] - new_df['standard_cost']
```

```python
new_df['total_profit'] =  new_df['profit'] * new_df['quantity']
```

```python
df_profit = new_df.groupby(['category']).agg({'total_profit':'sum'}).reset_index()
```

```python
df_profit
```

```python
# barplot
plt.figure(figsize=(10, 3))
sns.barplot(x = 'category',
            y = 'total_profit',
            hue = 'category',
            data = df_profit.sort_values(by ='total_profit', ascending=False))
plt.xlabel("Product Category")
plt.ylabel("Total Profit")
plt.title("Total Profit by Product Category")
plt.show()
```
